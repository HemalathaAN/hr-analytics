## 📦 Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8')
sns.set_palette('husl')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12

print('✅ Libraries imported successfully!')
```

## 📂 Step 2: Load & Explore Dataset

```python
# Load dataset
# Download from: https://www.kaggle.com/datasets/pavansubhasht/ibm-hr-analytics-attrition-dataset
df = pd.read_csv('hr_data.csv')

print('Dataset Shape:', df.shape)
print('\nColumns:', df.columns.tolist())
print('\nMissing Values:', df.isnull().sum().sum())
print('\nAttrition Distribution:')
print(df['Attrition'].value_counts())
df.head()
```

## 🧹 Step 3: Data Cleaning & Preprocessing

```python
# Drop useless columns
df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1, inplace=True)

# Convert Attrition to binary
df['Attrition_Binary'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Salary bands
df['Salary_Band'] = pd.cut(df['MonthlyIncome'],
    bins=[0, 3000, 6000, 10000, 20000],
    labels=['Low', 'Medium', 'High', 'Very High']
)

# Age groups
df['Age_Group'] = pd.cut(df['Age'],
    bins=[18, 25, 35, 45, 60],
    labels=['18-25', '26-35', '36-45', '46-60']
)

print('✅ Data cleaned and features engineered!')
print(f'Dataset shape: {df.shape}')
```

## 📊 Step 4: Exploratory Data Analysis

```python
# Overall Attrition Rate
attrition_rate = df['Attrition_Binary'].mean() * 100
print(f'📊 Overall Attrition Rate: {attrition_rate:.1f}%')

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Pie chart
attrition_counts = df['Attrition'].value_counts()
axes[0].pie(attrition_counts, labels=['Stayed', 'Left'],
            autopct='%1.1f%%', colors=['#0077b6', '#ff6b6b'], startangle=90)
axes[0].set_title('Overall Attrition Rate', fontweight='bold', fontsize=14)

# Bar chart
attrition_counts.plot(kind='bar', ax=axes[1], color=['#0077b6', '#ff6b6b'])
axes[1].set_title('Attrition Count', fontweight='bold', fontsize=14)
axes[1].set_xlabel('Attrition')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.savefig('attrition_overview.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Department wise Attrition
dept_attrition = df.groupby('Department')['Attrition_Binary'].mean() * 100
dept_attrition = dept_attrition.sort_values(ascending=False)

plt.figure(figsize=(10, 5))
dept_attrition.plot(kind='bar', color=['#ff6b6b', '#ffd93d', '#0077b6'])
plt.title('🏢 Attrition Rate by Department (%)', fontweight='bold', fontsize=14)
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.tick_params(axis='x', rotation=0)
for i, v in enumerate(dept_attrition):
    plt.text(i, v + 0.3, f'{v:.1f}%', ha='center', fontweight='bold')
plt.tight_layout()
plt.savefig('dept_attrition.png', dpi=150, bbox_inches='tight')
plt.show()
print(dept_attrition)
```

```python
# Salary vs Attrition
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Box plot
df.boxplot(column='MonthlyIncome', by='Attrition', ax=axes[0],
           boxprops=dict(color='#0077b6'),
           medianprops=dict(color='#ff6b6b', linewidth=2))
axes[0].set_title('💰 Monthly Income vs Attrition', fontweight='bold')
axes[0].set_xlabel('Attrition')
axes[0].set_ylabel('Monthly Income')

# Salary band attrition
salary_attrition = df.groupby('Salary_Band')['Attrition_Binary'].mean() * 100
salary_attrition.plot(kind='bar', ax=axes[1], color=['#ff6b6b', '#ffd93d', '#0077b6', '#00b4d8'])
axes[1].set_title('💵 Attrition Rate by Salary Band (%)', fontweight='bold')
axes[1].set_xlabel('Salary Band')
axes[1].set_ylabel('Attrition Rate (%)')
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.savefig('salary_attrition.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Overtime vs Attrition
overtime_attrition = df.groupby('OverTime')['Attrition_Binary'].mean() * 100

plt.figure(figsize=(8, 5))
overtime_attrition.plot(kind='bar', color=['#0077b6', '#ff6b6b'])
plt.title('⏰ Attrition Rate by Overtime (%)', fontweight='bold', fontsize=14)
plt.xlabel('Overtime')
plt.ylabel('Attrition Rate (%)')
plt.tick_params(axis='x', rotation=0)
for i, v in enumerate(overtime_attrition):
    plt.text(i, v + 0.5, f'{v:.1f}%', ha='center', fontweight='bold')
plt.tight_layout()
plt.savefig('overtime_attrition.png', dpi=150, bbox_inches='tight')
plt.show()
print(overtime_attrition)
```

```python
# Job Satisfaction vs Attrition
satisfaction_attrition = df.groupby('JobSatisfaction')['Attrition_Binary'].mean() * 100

plt.figure(figsize=(8, 5))
satisfaction_attrition.plot(kind='bar', color=['#ff6b6b', '#ffd93d', '#0077b6', '#00b4d8'])
plt.title('😊 Attrition Rate by Job Satisfaction', fontweight='bold', fontsize=14)
plt.xlabel('Job Satisfaction (1=Low, 4=High)')
plt.ylabel('Attrition Rate (%)')
plt.tick_params(axis='x', rotation=0)
plt.tight_layout()
plt.savefig('satisfaction_attrition.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Age Group vs Attrition
age_attrition = df.groupby('Age_Group')['Attrition_Binary'].mean() * 100

plt.figure(figsize=(10, 5))
age_attrition.plot(kind='bar', color=['#ff6b6b', '#ffd93d', '#0077b6', '#00b4d8'])
plt.title('🎂 Attrition Rate by Age Group (%)', fontweight='bold', fontsize=14)
plt.xlabel('Age Group')
plt.ylabel('Attrition Rate (%)')
plt.tick_params(axis='x', rotation=0)
for i, v in enumerate(age_attrition):
    plt.text(i, v + 0.3, f'{v:.1f}%', ha='center', fontweight='bold')
plt.tight_layout()
plt.savefig('age_attrition.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Correlation Heatmap
numeric_cols = df.select_dtypes(include=[np.number]).columns
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm',
            center=0, linewidths=0.5)
plt.title('🔥 Correlation Heatmap', fontweight='bold', fontsize=14)
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```
